## üßπ Data Cleaning

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
print(df.columns)
```

## üìà Extracting Population of 2021

```python
#Plotting the top 10 countries by population in 2021
df_2021 = df[['Country Name', '2021']]
df_2021 = df_2021.dropna()
df_2021.columns = ['Country', 'Population']
```

```python
print("‚úÖ Extracted 2021 Population Data (first 10 rows):")
display(df_2021.head(10))  # Display first 10 rows of cleaned data
```

## üåü Top 10 Countries by Population in 2021

```python
# Get Top 10 Most Populated Countries in 2021
top10 = df_2021.sort_values('Population', ascending=False).head(10)
```

```python
print("\nüåç Top 10 Countries by Population in 2021:")
display(top10)
```

## üìä Data Visualization - Bar Chart

```python
#  Plot bar chart for top 10 countries
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Country", y="Population", palette="viridis")
# Add titles and labels
plt.title("Top 10 Countries by Population in 2021", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
# Show the plot
plt.show()
```

## üìä Data Visualization - Histogram

```python
# Drop missing population data for 2021 (if any)
df_2021 = df.dropna(subset=['2021'])

plt.figure(figsize=(12, 6))
sns.histplot(df_2021['2021'], bins=30, kde=True, color='skyblue')

plt.title("Distribution of Country Populations in 2021", fontsize=16)
plt.xlabel("Population", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.tight_layout()
plt.show()
```
