## üßπ Data Preprocessing

```python
# Check for missing values
missing_values = df_sample.isnull().sum()
print("Missing values per column:\n", missing_values)
```

## üåç Handling Missing Values

```python
# Drop columns with too many missing values (>50%) [Optional]
threshold = 0.5 * len(df_sample)
df_cleaned = df_sample.dropna(thresh=threshold, axis=1)
```

```python
# Fill remaining missing values (for selected columns)

df_cleaned.loc[:, 'Weather_Condition'] = df_cleaned['Weather_Condition'].fillna('Unknown')
df_cleaned.loc[:, 'Visibility(mi)'] = df_cleaned['Visibility(mi)'].fillna(df_cleaned['Visibility(mi)'].median())
```

```python
df_cleaned = df_cleaned.copy()

df_cleaned['Start_Time'] = pd.to_datetime(df_cleaned['Start_Time'], errors='coerce')
df_cleaned['Hour'] = df_cleaned['Start_Time'].dt.hour
df_cleaned['Day'] = df_cleaned['Start_Time'].dt.day
df_cleaned['Weekday'] = df_cleaned['Start_Time'].dt.day_name()
```

```python
print("Cleaned data shape:", df_cleaned.shape)
df_cleaned.head()
```

## üîç Exploratory Data Analysis (EDA)

## ü•ß Accidents by Severity ‚Äî Pie Chart




```python
# Count accidents by severity
accidents_severity = df_cleaned['Severity'].value_counts().sort_index()

print(accidents_severity)
```

```python
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
labels = ['Severity 1', 'Severity 2', 'Severity 3', 'Severity 4']
# Use bright color palette for the pie
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC66']
wedges, texts, autotexts = ax.pie(
    accidents_severity,
    labels=labels,
    autopct='%1.1f%%',
    pctdistance=0.85,
    colors=colors,
    textprops={'fontsize': 12}
)
# Draw center circle for donut shape
circle = plt.Circle((0, 0), 0.5, color='white')
ax.add_artist(circle)
ax.set_title("Accidents by Severity", fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.tight_layout()
plt.show()
```

## ‚è∞ Accidents by Time of Day (Hour) ‚Äî Histogram

```python
plt.figure(figsize=(10,5))
sns.histplot(df_cleaned['Hour'], bins=24, kde=True, color='skyblue')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()
```

## üìÖ Accidents by Weekday ‚Äî Barplot

```python
plt.figure(figsize=(8, 6))
sns.countplot(
    data=df_cleaned,
    x='Weekday',
    hue='Weekday',  # Required to safely use `palette`
    palette='bright',  # Bright color palette
    legend=False  # Avoid unnecessary legend
)
plt.title("Accidents by Day of Week")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üå¶Ô∏è Top 10 Weather Conditions During Accidents ‚Äî Barplot



```python
plt.figure(figsize=(10,5))
top_weather = df_cleaned['Weather_Condition'].value_counts().head(10)
sns.barplot(
    x=top_weather.index,
    y=top_weather.values,
    hue=top_weather.index,
    palette='Dark2',
    legend=False
)
plt.title('Top 10 Weather Conditions')
plt.xticks(rotation=45)
plt.show()
```
